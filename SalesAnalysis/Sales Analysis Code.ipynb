# Sales Analysis

#### Import Necessary Libraries

```python
import pandas as pd
import os
```

#### Merge the 12 months of sales data into a single CSV file

```python
df = pd.read_csv("./Sales Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales Data')]

all_months_data = pd.DataFrame()

for file in files:
    df =  pd.read_csv("./Sales Data/"+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv", index=False)
    
```

#### Read in upadted dataframe

```python
 all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Clean Up the Data

#### Drop rows of NAN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data =  all_data.dropna(how='all')
all_data.head()
```

### Find 'Or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

### Convert Columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #Make float
```

### Augment Data with Additional Columns

#### Task 2: Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Task 3: Add a sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Task 4: Add a city column

```python
# Let's use .apply()    
def get_city(address):
    return address.split(',')[1]
    
def get_state(address):
    return address.split(',')[2].split(' ')[1]
    
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()


#all_data = all_data.drop(columns='Column')
```

### Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

### To visualize we plot a graph to kind of see the trend

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

## Question 4: What products are most often sold together

```python
# The hint in this case is that products that shares the same order number means that they were sold together.

df = all_data[all_data['Order ID'].duplicated(keep=False)]

df ['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

 ##### Task: Now we want to count pairs of the grouped item

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
      print(key, value)
 
```

```python
#hint 
#we need to sum up the Quantity Ordered based on groupby product

import os
import pandas as pd

# Assuming you have already loaded and preprocessed the data as 'all_data'

# Group by product and sum the quantity ordered
product_group = all_data.groupby('Product')
quantity_ordered_sum = product_group['Quantity Ordered'].sum()

# Find the product that sold the most
most_sold_product = quantity_ordered_sum.idxmax()
most_sold_quantity = quantity_ordered_sum.max()

# Print the result
print(f"The product that sold the most is: {most_sold_product}")
print(f"Total quantity sold: {most_sold_quantity}")

# Optional: Analysis - More insights based on the data

# Let's look at the sales for the most sold product
most_sold_product_sales = all_data[all_data['Product'] == most_sold_product]['Sales'].sum()
print(f"Total sales for the most sold product: ${most_sold_product_sales:.2f}")

# Plotting the quantities for visualization (optional)
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
quantity_ordered_sum.plot(kind='bar')
plt.title('Total Quantity Ordered by Product')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=90)
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded and preprocessed the data as 'all_data'

# Group by product and sum the quantity ordered
product_group = all_data.groupby('Product')
quantity_ordered_sum = product_group['Quantity Ordered'].sum()

# Find the product that sold the most
most_sold_product = quantity_ordered_sum.idxmax()
most_sold_quantity = quantity_ordered_sum.max()

# Print the result
print(f"The product that sold the most is: {most_sold_product}")
print(f"Total quantity sold: {most_sold_quantity}")

# Optional: Analysis - you can add more insights based on the data
# For example, let's look at the sales for the most sold product
most_sold_product_sales = all_data[all_data['Product'] == most_sold_product]['Sales'].sum()
print(f"Total sales for the most sold product: ${most_sold_product_sales:.2f}")

# Average price of each product
average_price = product_group['Price Each'].mean()

# Plotting the quantities and prices for visualization
fig, ax1 = plt.subplots(figsize=(10,6))

ax1.bar(quantity_ordered_sum.index, quantity_ordered_sum.values, color='b', label='Quantity Ordered')
ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_xticklabels(quantity_ordered_sum.index, rotation=90)

# Create a second y-axis
ax2 = ax1.twinx()
ax2.plot(quantity_ordered_sum.index, average_price.values, color='r', label='Average Price', marker='o')
ax2.set_ylabel('Price Each ($)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Adding title and legend
fig.suptitle('Total Quantity Ordered and Average Price by Product')
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

plt.show()
```
